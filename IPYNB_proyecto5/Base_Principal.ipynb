```python
import osmnx as ox
import pandas as pd

# 1. Descarga del área de interés
G = ox.graph_from_address("Museo Soumaya, Ciudad de México", dist=2000, network_type="bike")

# 2. Convertir a GeoDataFrames
nodes, edges = ox.graph_to_gdfs(G)

# 3. Resetear el índice para convertir u, v y key en columnas normales
edges = edges.reset_index()

def generar_base_abt(edges_df, nodes_df):
    # Crear una copia para no alterar el original
    abt = edges_df.copy()

    # --- Identificadores y Nombres ---
    # Creamos un segment_id único combinando los nodos
    abt['segment_id'] = abt['u'].astype(str) + "_" + abt['v'].astype(str) + "_" + abt['key'].astype(str)
    abt['street_name'] = abt.get('name', 'Sin nombre')
    abt['alcaldia_name'] = "Miguel Hidalgo" # En este caso es manual, para masivo usaríamos un Spatial Join

    # --- Coordenadas de Inicio y Fin ---
    # Mapeamos los nodos 'u' (inicio) y 'v' (fin) con sus lat/lon de la tabla nodes
    abt['lat_start'] = abt['u'].map(nodes_df['y'])
    abt['lon_start'] = abt['u'].map(nodes_df['x'])
    abt['lat_end'] = abt['v'].map(nodes_df['y'])
    abt['lon_end'] = abt['v'].map(nodes_df['x'])

    # --- Atributos de Infraestructura ---
    # Usamos .get() para evitar errores si la columna no existe en esta zona de OSM
    abt['highway'] = abt.get('highway', 'unclassified')
    abt['lanes'] = abt.get('lanes', 1)
    abt['maxspeed'] = abt.get('maxspeed', '30')
    abt['oneway'] = abt.get('oneway', False)
    abt['length'] = abt.get('length', 0)
    abt['junction'] = abt.get('junction', 'no')
    abt['bridge'] = abt.get('bridge', 'no')
    abt['width'] = abt.get('width', None)
    abt['service'] = abt.get('service', None)
    abt['access'] = abt.get('access', 'public')

    # --- Variables de Interés Ciclista (Ciclovía e Iluminación) ---
    # Convertimos a formato binario o categórico limpio
    abt['ciclovia'] = abt.apply(lambda x: 1 if 'cycleway' in x or 'bicycle' in x else 0, axis=1)
    abt['iluminacion'] = abt.get('lit', 'no')

    # --- Selección y Orden de Columnas ---
    columnas_finales = [
        'segment_id', 'street_name', 'alcaldia_name', 'highway', 'lanes',
        'maxspeed', 'oneway', 'length', 'geometry', 'lat_start',
        'lon_start', 'lat_end', 'lon_end', 'junction', 'bridge',
        'width', 'service', 'access', 'ciclovia', 'iluminacion'
    ]

    return abt[columnas_finales]

# Ejecutar transformación
df_abt_base = generar_base_abt(edges, nodes)

# Ver resultado
df_abt_base.head()
```

# Codigo Generalizado para extraer toda la CDMX

```python
import osmnx as ox

# 1. Configuración de "Poder": Aumentamos el tiempo de espera y el tamaño de la memoria
ox.settings.timeout = 1000  # segundos
ox.settings.max_query_area_size = 2500000000 # bytes

print("🚀 Iniciando descarga de toda la CDMX (esto puede tardar 5-10 min)...")

# 2. Descargar la red completa de CDMX
# Usamos el nombre oficial para que encuentre el polígono completo
try:
    G_cdmx = ox.graph_from_place("Ciudad de México, México", network_type="bike")

    # 3. Convertir a GeoDataFrames
    nodes, edges = ox.graph_to_gdfs(G_cdmx)

    print(f"✅ Éxito. Nodos: {len(nodes)}, Segmentos (Edges): {len(edges)}")
except Exception as e:
    print(f"❌ Error: {e}")
```

```python
import pandas as pd
import geopandas as gpd
import os

# 1. Aplicar limpieza a TODAS las columnas excepto geometry
def limpiar_todo(df):
    df_clean = df.copy()
    for col in df_clean.columns:
        if col != 'geometry':
            # Si hay listas, las une con ";"
            df_clean[col] = df_clean[col].apply(lambda x: "; ".join(map(str, x)) if isinstance(x, list) else x)
            # Asegurar que todo sea string o numérico, nada de objetos raros
            if df_clean[col].dtype == 'object':
                df_clean[col] = df_clean[col].astype(str).replace('None', None).replace('nan', None)
    return df_clean

df_abt_final = limpiar_todo(df_abt_base)

# 2. El truco de la Geometría:
# Si usas to_parquet de PANDAS, la geometría da error.
# Debes usar GEOPANDAS para guardar como GeoParquet.
folder_path = '/content/drive/MyDrive/Colab Notebooks/Bases_datos_Proy5'
file_name = 'ABT_raiz_red_ciclista_completa_cdmx.parquet'
# full_path = os.path.join(folder_path, file_name)
full_path_geoparquet = os.path.join(folder_path, file_name)

# Convertir a GeoDataFrame por si acaso
if not isinstance(df_abt_final, gpd.GeoDataFrame):
    df_abt_final = gpd.GeoDataFrame(df_abt_final, geometry='geometry')

# Guardar (Esto sí acepta la columna geometry correctamente)
df_abt_final.to_parquet(full_path_geoparquet)

print(f"✅ Guardado como GeoParquet en: {full_path_geoparquet}")
```
